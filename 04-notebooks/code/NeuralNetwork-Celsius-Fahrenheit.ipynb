```python
import tensorflow as tf
import numpy as np

# Datos de entrenamiento
celsius = np.array([-40, -10, 0, 8, 15, 22, 38], dtype=float)
fahrenheit = np.array([-40, 14, 32, 46, 59, 72, 100], dtype=float)
```

```python
# Capa densa con 1 sola neurona
capa = tf.keras.layers.Dense(units=1, input_shape=[1])
#                            neuronas, cantidad de inputs (1 porque solo se ingresa un valor, si fuera una imagen por ejemplo serГ­a (128, 128))
modelo = tf.keras.models.Sequential([capa])
```

```python
# CompilaciГіn
modelo.compile(optimizer=tf.keras.optimizers.Adam(0.1), loss='mean_squared_error')

# Entrenamiento
modelo.fit(celsius, fahrenheit, epochs=500, verbose=True)
# modelo.fit(caracterГ­sticas, etiquetas, repeticiones (epocas), barra de progreso por epoca (0|1|2) (true/false))
```

```python
# Obtener los valores internos de la capa
pesos = capa.get_weights()

# pesos[0] es la matriz de pesos (W)
# pesos[1] es el arreglo de sesgos (b)
print(f"Variables internas del modelo:")
print(f"Peso (Weight): {pesos[0][0][0]}")
print(f"Sesgo (Bias): {pesos[1][0]}")
```

```python
# 1. Supongamos que ya tienes el modelo entrenado
# 2. Creamos los nuevos datos
celsius_test = np.array([0, 15, 38], dtype=float)
 
# 3. Predecimos
resultados = modelo.predict(celsius_test)
 
# 4. Mostramos los resultados de forma limpia
print("--- Predicciones del Modelo ---")
# resultados[0] serГЎ la predicciГіn para 0В°C
# resultados[1] serГЎ la predicciГіn para 15В°C, etc.
print(resultados)
```
